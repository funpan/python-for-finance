### Import and plot stock price data with python, pandas and seaborn

###### Date: 2017-03-19
###### Ref: http://www.augustkleimo.com/import-and-plot-stock-price-data-with-python-pandas-and-seaborn/

```python
import numpy as np
import matplotlib.pyplot as pp
import pandas as pd
import seaborn as sns
import urllib.request
```

```python
# Instruct Python to show our plots inline on the screen.
%matplotlib inline
```

```python
# Fetch .csv data file from yahoo finance
urllib.request.urlretrieve('http://real-chart.finance.yahoo.com/table.csv?s=SPY&d=18&e=3&f=2017&g=d&a=0&b=29&c=1993',
    'spy.csv')
```

```python
open('spy.csv', 'r').readlines()[:10]
```

```python
spy = pd.read_csv("spy.csv", parse_dates = ['Date'])
```

```python
spy.dtypes
```

```python
# sort order using sort_values() function.
spy = spy.sort_values(by='Date')
```

```python
# Set the trading date as index using set_index() function.
spy.set_index('Date', inplace=True)
```

```python
# Plot the closig price of SPY over the entire period
spy['Close'].plot(figsize=(16,12))
```

```python
# Use the truncate() function to remove data prior to Jan 1st 2015
spy.truncate(before="2015-01-01")['Close'].plot(figsize=(16,12))
```
